```python
# Dataframe with functions

import numpy as np
import pandas as pd

df = pd.DataFrame(np.arange(16).reshape(4,4),index=list('abcd'),columns=['Pune','Mumbai','Chennai','Delhi'])
print("*"*60)
print(df)
print("*"*60)
f = lambda x: x.max() - x.min()
df.apply(f)
```

```python
# Dataframe with series function

import numpy as np
import pandas as pd

def f(x):
    return pd.Series([x.min(),x.max()],index=['min','max'])


df = pd.DataFrame(np.arange(16).reshape(4,4),index=list('abcd'),columns=['Pune','Mumbai','Chennai','Delhi'])
print("*"*60)
print(df)
print("*"*60)
df.apply(f)
```

```python
# Sorting Data Frame

import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(4,4).reshape(4,4),index=list('acbd'),columns=['Pune','Mumbai','Chennai','Delhi'])
print("*"*60)
print(df)
print("*"*60)
print("\n","*"*60)
print(df.sort_index(axis=0))
print("*"*60)
print("\n","*"*60)
print(df.sort_index(axis=1,ascending=False))
print("*"*60,"\n")
```
